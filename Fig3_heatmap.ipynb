```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
#  Create heatmaps for Emax and pEC50 double normalised data
def create_heatmap(input_sheet = 'Emax_dnorm'):
    df_plot = pd.read_excel('data/data.xls', sheet_name=input_sheet, engine='xlrd', skiprows=4)

    df_plot = df_plot.fillna(0)
    df_plot.index = df_plot['Receptor']
    df_plot = df_plot[[col for col in df_plot.columns if col not in ['Receptor', 'Ligand']]]

    plt.clf()
    sns.set(font_scale=0.4, rc={'figure.figsize': (4, 16)})
    cmap = 'RdYlBu_r'

    # Mask NANs
    cm = sns.clustermap(df_plot, row_cluster=False, cmap=cmap, mask=df_plot == 0, col_cluster=False, linewidths=0.1, xticklabels=True, yticklabels=True)

    cm.cax.set_position([.1, .1, .02, .45])
    col = cm.ax_col_dendrogram.get_position()
    hm = cm.ax_heatmap.get_position()
    cm.ax_heatmap.set_position([hm.x0, hm.y0, hm.width * 0.25, hm.height])
    cm.ax_col_dendrogram.set_position([col.x0, col.y0, col.width * 0.25, col.height * 0.5])

    plt.show()
```

```python
create_heatmap('Emax_dnorm')
```

```python
create_heatmap('pEC50_dnorm')
```
